Para este projeto trabalhamos com uma base de dados com números sobre incêndios no Brasil. Pudemos ver quais estados mais sofrem com as queimadas e qual período teve mais queimadas no Brasil.

```python
# Libs para Modelagem e Matrizez
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
# Lendo os dados
Base_Dados = pd.read_csv('Dados_Indendio.csv', encoding='latin-1')

# Verificando
Base_Dados.head() 
```

```python
# Vendo o tamanho da base de dados
Base_Dados.shape
```

```python
# Vendo se tem dados nulos
Base_Dados.isnull().sum()
```

```python
# Algumas informações da basede dados
Base_Dados.info()
```

```python
# Vendo estatísticas da base de dados
Base_Dados.describe()
```

```python
# Vendo os dados únicos
Base_Dados.nunique()
```

```python
# Agrupando dados por ano
Analise = Base_Dados.groupby( by=['year'] ).sum().reset_index()
Analise.head()
```

```python
# Gerando um gráfico de incêndios por ano
plt.figure( figsize=(12, 5) )
plt.style.use('ggplot')

plt.title( 'Indêncidios no Brasil: 1997 - 2017', loc='left', fontsize=14 )
sns.lineplot( data=Analise, x='year', y='number', estimator='sum', lw=2, color='#ff5555', alpha=0.85 );

# Nomeando as labels
plt.xlabel('Quantidade de incêndios')
plt.ylabel('Período');
```

```python
# Agrupando por mês
Analise_02 = Base_Dados.groupby( by=['year', 'month'] ).sum().reset_index()
Analise_02.head()
```

```python
# Analise de incêndios por ano
plt.figure( figsize=(12, 5) )
plt.title( 'Incêndios por mês', loc='left', fontsize=14 )
sns.boxplot( data=Analise_02, x='month', y='number', palette='coolwarm', saturation=1, width=0.9, linewidth=2,
            order=['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro'] )

# Nomeando as labels
plt.xlabel('Mês')
plt.ylabel('Número de incêndios');
```

```python
# Agrupando por estados e plotando um gráfico de barras
Analise_03 = Base_Dados.groupby( by=['state'] ).sum().reset_index()[['state', 'number']].sort_values( 'number', ascending=False )
Analise_03.head()

plt.figure( figsize=(12, 5) )
plt.title( 'Estados com maior número de incêndios', loc='left', fontsize=14 )
plt.bar( Analise_03.state, Analise_03['number'], color='#f44e3f' )

# Nomeando as labels
plt.ylabel('Quantidade de incêncios')
plt.xticks( rotation=90 );
```

```python
# Estados TOP 10 com mais incêndios
Lista_TOP10 = Analise_03['state'][0:10].values

# Tamanho
plt.figure( figsize=(12, 5) )

# Gráfico
plt.title( 'Top 10 Estados com incêncios', loc='left', fontsize=14 )

# Loop
for Coluna in Lista_TOP10:
  
# Filtrar o estado
  Filtro = Base_Dados.loc[ Base_Dados['state'] == Coluna ]

# Agrupar os valores para sumarizar
  Analise_Local = Filtro.groupby( by=['year'] ).sum().reset_index()

# Plot
  sns.lineplot(data=Analise_Local, x='year', y='number', lw=2, alpha=0.85)

# Nomeando as Labels
plt.xlabel('Período')
plt.ylabel('Número de incêndios')

# Legenda
plt.legend( Lista_TOP10, bbox_to_anchor=(1, 0.7) );
```

```python
# Adicionando dados de geolocalização para plotagem de gráfico

# Gerando os estados
Estados = Analise_03.sort_values('state')['state'].values

# Gerando os valores
Valores = Analise_03.sort_values('state')['number'].values

# Adicionando as Latitudes
Lat = [ -8.77, -9.71,	1.41, -3.07,	-12.96, -3.71, -15.83, -19.19, -16.64, -2.55,	-12.64,	
       -18.10, -7.06, -5.53, -8.28, -8.28,	-22.84,	-11.22,	1.89,	-27.33,	-23.55,	-10.90,	-10.25 ]

# adicionando as Longitudes
Long = [ -70.55,	-35.73,	-51.77,	-61.66,	-38.51,	-38.54,	-47.86,	-40.34,	-49.31,	-44.30,	-55.42,	-44.38,	
       -35.55,	-52.29,	-35.07,	-43.68,	-43.15,	-62.80,	-61.22,	-49.44,	-46.64,	-37.07,	-48.25 ]

# Organizando os dados
Dicionario = {
    'Estados' : Estados,
    'Latitude' : Lat,
    'Longitude' : Long,
    'Incêndios' : Valores
}

# Lendo o dicionario
Analise_Geografica = pd.DataFrame ( Dicionario )

Analise_Geografica.head()
```

```python
# Importando a biblioteca e plotando o gráfico
import plotly.express as px

# Mapa de Calor Geografico
px.density_mapbox(
    Analise_Geografica,
    lat='Latitude',
    lon='Longitude',
    z='Incêndios',
    radius=30,
    center=dict(lat=-12.700, lon=-46.5555),
    zoom=3,
    mapbox_style='stamen-terrain'
)
```
